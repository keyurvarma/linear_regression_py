```python
# Importing the necessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plotpy
plotpy.rcParams['figure.figsize'] = (10.0, 7.0)

#function to calculate m 
def getM(X,Y,X_mean,Y_mean):
    num = 0
    den = 0
    for i in range(len(X)):
        num += (X[i] - X_mean)*(Y[i] - Y_mean)
        den += (X[i] - X_mean)**2
    m = num / den
    return m

# Preprocessing the Input data from Training Dataset
training_data = pd.read_csv(R"T:\Python\dataset\train.csv")
X = training_data.iloc[:, 0]
Y = training_data.iloc[:, 1]
plotpy.scatter(X, Y)
plotpy.show()

# Preprocessing the Input data from the Ideal Functions Dataset
ideal_data = pd.read_csv(R"T:\Python\dataset\ideal.csv")
X = ideal_data.iloc[:, 0]
Y = ideal_data.iloc[:, 1]
plotpy.scatter(X, Y)
plotpy.show()

# Building the model
X_mean = np.mean(X)
Y_mean = np.mean(Y)

m = getM(X, Y, X_mean, Y_mean)
c = Y_mean - m*X_mean

#displaying values of m & c
print (m, c)

# Making predictions
Y_pred = m*X + c

plotpy.scatter(X, Y) # actual
# Red color is used for the predicted regression line
plotpy.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red') #predicted
plotpy.show()
```
